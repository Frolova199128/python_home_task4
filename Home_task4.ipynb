## Задание 3

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3]=0
corr_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, linewidths=1, cmap='coolwarm');
```

```python
## Видим сильную положитеьную корреляция между показателями цена и жилая площадь, цена и дизайн, цена и этажность, 
```

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.50);
plt.xlabel('цена')
plt.ylabel('вид из окна')
plt.title('Соотноение стоимости домой и вида за окном');
```

```python
## Самая высокая стоимость у домой с оценков вида 4
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.50);
plt.xlabel('цена')
plt.ylabel('количество этажей')
plt.title('Соотноение стоимости домой и этажности');
```

```python
## Самая высокая стоимость у домов с 2,5 этажами
```

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.50);
plt.xlabel('цена')
plt.ylabel('оценка дизайна')
plt.title('Соотноение стоимости домой и дизайна');
```

```python
## Самыми дорогими являются дома с оценкой конструкция и дизайна 13
```
